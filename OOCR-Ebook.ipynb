```python
from reportlab.pdfgen import canvas
from reportlab.lib.pagesizes import letter
from reportlab.pdfbase.ttfonts import TTFont
from reportlab.pdfbase import pdfmetrics

def create_pdf(text, output_path):
    # Register Thai font
    pdfmetrics.registerFont(TTFont('THSarabunNew', '/Users/apichet/THSarabunNew.ttf'))
    
    my_canvas = canvas.Canvas(output_path, pagesize=letter)
    my_canvas.setFont("THSarabunNew", 12)
    
    # Split text into lines to fit the page
    lines = text.split("\n")
    y = 750  # Start y position for drawing text
    
    for line in lines:
        if y < 40:  # Move to next page if not enough space
            my_canvas.showPage()
            my_canvas.setFont("THSarabunNew", 12)
            y = 750
        
        my_canvas.drawString(40, y, line)
        y -= 20  # Move y position down for the next line
    
    my_canvas.save()

# Example usage
text_content = """
สวัสดีครับ นี่คือตัวอย่างข้อความภาษาไทย
ที่จะถูกสร้างเป็นไฟล์ PDF ด้วยฟอนต์ไทย
"""
pdf_output_path = '/Users/apichet/ebook.pdf'
create_pdf(text_content, pdf_output_path)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from PIL import Image, ImageEnhance, ImageFilter
import pytesseract
from io import BytesIO
import time
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas
from reportlab.pdfbase.ttfonts import TTFont
from reportlab.pdfbase import pdfmetrics
import os

# Path to Thai font file (update this path to the correct path)
thai_font_path = '/Users/apichet/THSarabunNew.ttf'  # Update to the actual path
output_dir = '/Users/apichet/Downloads/ebook_pages3'  # Directory to save OCR text files

# Ensure the output directory exists
os.makedirs(output_dir, exist_ok=True)

# Initialize Selenium WebDriver
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# URL of the eBook
url = "https://www.tgia.org/E-book/MotorInsPolicyInterpret2020/mobile/index.html"
driver.get(url)

time.sleep(5)  # Wait for the page to load

# Check for the correct selector for the "Next" button
try:
    next_button_xpath = '//*[@class="flip_button_right"]'  # Update with the actual XPath if necessary
    next_button = driver.find_element(By.XPATH, next_button_xpath)
except Exception as e:
    print(f"Error finding the next button: {e}")

# Iterate over all pages
for i in range(120):  # Assuming there are 120 pages
    try:
        # Screenshot the current page
        screenshot = driver.get_screenshot_as_png()
        image = Image.open(BytesIO(screenshot))

        # Enhance image quality for OCR
        image = image.convert('L')  # Convert to grayscale
        image = ImageEnhance.Contrast(image).enhance(2)  # Increase contrast
        image = image.filter(ImageFilter.SHARPEN)  # Sharpen the image
        image = image.filter(ImageFilter.MedianFilter())  # Apply median filter
        
        # Use OCR to extract text
        custom_oem_psm_config = r'--oem 3 --psm 1'
        text = pytesseract.image_to_string(image, lang='tha', config=custom_oem_psm_config)
        
        # Save text to a file
        with open(os.path.join(output_dir, f'page_{i+1}.txt'), 'w', encoding='utf-8') as f:
            f.write(text)
        
        # Go to the next page
        next_button = driver.find_element(By.XPATH, next_button_xpath)
        next_button.click()
        time.sleep(1)  # Wait for the next page to load
    except Exception as e:
        print(f"Error on page {i + 1}: {e}")
        break

# Close the browser
driver.quit()

# Combine all text files into one
full_text = ""
for i in range(120):
    try:
        with open(os.path.join(output_dir, f'page_{i+1}.txt'), 'r', encoding='utf-8') as f:
            full_text += f.read() + "\n\n"
    except FileNotFoundError:
        break

# Create PDF from text with Thai font
def create_pdf(text, output_path):
    pdfmetrics.registerFont(TTFont('THSarabunNew', thai_font_path))  # Path to Thai font file
    my_canvas = canvas.Canvas(output_path, pagesize=letter)
    my_canvas.setFont("THSarabunNew", 12)
    
    # Split text into lines to fit the page
    lines = text.split("\n")
    y = 750  # Start y position for drawing text
    
    for line in lines:
        if y < 40:  # Move to next page if not enough space
            my_canvas.showPage()
            my_canvas.setFont("THSarabunNew", 12)
            y = 750
        
        my_canvas.drawString(40, y, line)
        y -= 20  # Move y position down for the next line
    
    my_canvas.save()

# Create PDF document
pdf_output_path = '/Users/apichet/ebook.pdf'
create_pdf(full_text, pdf_output_path)

pdf_output_path
```
